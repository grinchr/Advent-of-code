```python
def map_up(index, n):
    row = index // n
    col = index % n
    return int(row), int(col)

def map_down(x, y, n):
    return int(x * n + y)

def insert_newline_every_n_chars(input_string, n):
    return '\n'.join(input_string[i:i+n] for i in range(0, len(input_string), n))

def update_map(input_string, tup, n):
    x,y = tup
    new_char = "#"
    index = map_down(x,y,n)
    return input_string[:index] + new_char + input_string[index + 1:]

test = False

if not test:
    data=real_data
    n=71
else:
    data=test_data
    n=7

grid = "."*n*n

for i, xy in enumerate(data):
    if i == 1024:
        break
    y,x = xy.split(",")
    tup = (int(x),int(y))
    if len(tup)!=2:
        print(tup)
    grid=update_map(grid,tup,n)


grid_print = insert_newline_every_n_chars(grid,n)
print(grid_print)
```

```python
def create_graph(grid):
    g = dict()

    for i,char  in enumerate(grid):

        if char!="#":
            x,y = map_up(i, n)
            g[(x,y)]=dict()
            #up
            if  x-1>=0 and grid[map_down(x-1, y, n)]!="#":
                g[(x,y)][(x-1,y)] = 1

            #left    
            if 0<=y-1 and grid[map_down(x, y-1, n)]!="#":
                g[(x,y)][(x,y-1)] = 1

            #right    
            if y+1<n and grid[map_down(x, y+1, n)]!="#":
                g[(x,y)][(x,y+1)] = 1
                
            #down
            if x+1<n and grid[map_down(x+1, y, n)]!="#":
                g[(x,y)][(x+1,y)] = 1
    return g
```

```python
import heapq

def dijkstra(g, source, target=None):

    # Priority queue for nodes to visit
    pq = []
    heapq.heappush(pq, (0, source))  # (distance, node)
    
    # Distance from source to each node
    distances = {node: float('inf') for node in g}
    distances[source] = 0

    # Previous node in the shortest path
    previous = {node: None for node in g}

    while pq:
        # Get the node with the smallest distance
        current_distance, current_node = heapq.heappop(pq)
        if current_distance > distances[current_node]:
            continue

        # Explore neighbors
        for neighbor, weight in g[current_node].items():
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                previous[neighbor] = current_node
                heapq.heappush(pq, (distance, neighbor))

    return distances, previous


src = (0,0)
target = (n-1,n-1)
g = create_graph(grid)
dist, prev = dijkstra(g, src)
```
