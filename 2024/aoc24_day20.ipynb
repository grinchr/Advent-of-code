```python

import os
file_path = f'input/day20.txt'
with open(file_path, 'r') as file:
    real_data = file.read()  # .split("\n")
```

```python
import os

#notebook_filename = os.path.basename(__file__) 

test_data = """###############
#...#...#.....#
#.#.#.#.#.###.#
#S#...#.#.#...#
#######.#.#.###
#######.#.#...#
#######.#.###.#
###..E#...#...#
###.#######.###
#...###...#...#
#.#####.#.###.#
#.#...#.#.#...#
#.#.#.#.#.#.###
#...#...#...###
###############"""


```

```python
from math import sqrt

def map_up(index, n):
    row = index // n
    col = index % n
    return int(row), int(col)

def map_down(x, y, n):
    return int(x * n + y)

def create_graph(grid,n):
    g = dict()

    for i,char  in enumerate(grid):

        if char!="#":
            x,y = map_up(i, n)
            g[(x,y)]=dict()
            #up
            if  x-1>=0 and grid[map_down(x-1, y, n)]!="#":
                g[(x,y)][(x-1,y)] = 1

            #left    
            if 0<=y-1 and grid[map_down(x, y-1, n)]!="#":
                g[(x,y)][(x,y-1)] = 1

            #right    
            if y+1<n and grid[map_down(x, y+1, n)]!="#":
                g[(x,y)][(x,y+1)] = 1
                
            #down
            if x+1<n and grid[map_down(x+1, y, n)]!="#":
                g[(x,y)][(x+1,y)] = 1
    return g

def insert_newline_every_n_chars(input_string, n):
    return '\n'.join(input_string[i:i+n] for i in range(0, len(input_string), n))

def update_map(input_string, x,y, n, char="0"):
    new_char = char
    index = map_down(x,y,n)
    return input_string[:index] + new_char + input_string[index + 1:]

data = real_data
data= data.replace("\n","")

n = int(sqrt(len(data)))

start_pos = map_up(data.find("S"),n)

target = map_up(data.find("E"),n)
```

```python
start_pos
target
g = create_graph(data, n)
```

```python
import heapq


src = start_pos
dist = {node: float("inf") for node in g}
prev = {node: None for node in g}

pq = list()
heapq.heappush(pq,(0, src))
dist[src] = 0
while pq:
    cur_distance, cur_node =  heapq.heappop(pq)
    if cur_distance > dist[cur_node]:
            continue
    
    for nabo, nabo_cost in g[cur_node].items():
        dist_to_nabo = cur_distance+nabo_cost
        if  dist[nabo] > dist_to_nabo:
            dist[nabo] = dist_to_nabo
            prev[nabo] = cur_node
            heapq.heappush(pq, (dist_to_nabo, nabo))
        
```

```python
target_found=False
last_step = target
steps = list()
while not target_found:
    steps.append(last_step)
    last_step = prev[last_step]
    if last_step==start_pos:
        target_found=True
steps.append(start_pos)
steps = list(reversed(steps))
```

```python

tally = {}

output_data = data
for i, xy in enumerate(steps):
    #up cheet
    x,y = xy
    if  x-2>=0 and data[map_down(x-1, y, n)]=="#" and data[map_down(x-2, y, n)]!="#":
        saved = dist[(x,y)] -dist[(x-2,y)] - 2
        if saved>0 :
            output_data = update_map(output_data, x-1,y,n,"1")
            output_data = update_map(output_data, x-2,y,n,"2")

            if saved in tally:
                tally[saved] += 1
            else:
                tally[saved] = 1

    #left    
    if 0<=y-2 and data[map_down(x, y-1, n)]=="#" and data[map_down(x, y-2, n)]!="#":
        saved = dist[(x,y)] -dist[(x,y-2)] - 2
        if saved>0:
            output_data = update_map(output_data, x,y-1,n,"1")
            output_data = update_map(output_data, x,y-2,n,"2")
            if saved in tally:
                tally[saved] += 1
            else:
                tally[saved] = 1

    #right    
    if y+2<n and data[map_down(x, y+1, n)]=="#" and data[map_down(x, y+2, n)]!="#":
        saved = dist[(x,y)] -dist[(x,y+2)] - 2
        if saved>0:
            output_data = update_map(output_data, x,y+1,n,"1")
            output_data = update_map(output_data, x,y+2,n,"2")
            if saved in tally:
                tally[saved] += 1
            else:
                tally[saved] = 1
            
    #down
    if x+2<n and data[map_down(x+1, y, n)]=="#" and data[map_down(x+2, y, n)]!="#":
        saved = dist[(x,y)] - dist[(x+2,y)] - 2
        if saved>0:
            output_data = update_map(output_data, x+1,y,n,"1")
            output_data = update_map(output_data, x+2,y,n,"2")
            if saved in tally:
                tally[saved] += 1
            else:
                tally[saved] = 1

```
