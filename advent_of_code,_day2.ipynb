```python
round_alter = {
    'A X' : 'A Z', 
    'B X' : 'B X',
    'C X' : 'C Y', 

    'A Y' : 'A X', 
    'B Y' : 'B Y', 
    'C Y' : 'C Z', 

    'A Z' : 'A Y', 
    'B Z' : 'B Z', 
    'C Z' : 'C X' 
}
score = np.vectorize(lambda a : scoremap[round_alter[a]])(rounds)
np.sum(score)
```
